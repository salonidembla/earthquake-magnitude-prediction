```python
# Import necessary libraries
import pandas as pd

# Load the dataset
file_path = 'filtered_earthquake_data.csv'  # Specify the correct path
earthquake_data = pd.read_csv(file_path)

# Display the first few rows to understand the structure of the dataset
earthquake_data.head()
```

```python
# Check column names
earthquake_data.columns
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plot the distribution of earthquake magnitudes
plt.figure(figsize=(10, 6))
sns.histplot(earthquake_data['source_magnitude'], bins=30, kde=True)
plt.title('Distribution of Earthquake Magnitudes')
plt.xlabel('Magnitude')
plt.ylabel('Frequency')
plt.show()
```

```python
# Select columns that are relevant for correlation analysis
seismic_factors = earthquake_data[['source_magnitude', 'source_depth_km', 
                                   'source_distance_km', 'snr_db']]

# Compute correlation matrix
correlation_matrix = seismic_factors.corr()

# Plot heatmap of correlation matrix
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Between Seismic Factors')
plt.show()
```

```python
# Plot the number of earthquakes over time
plt.figure(figsize=(10, 6))
earthquake_data.groupby('origin_year').size().plot(kind='bar')
plt.title('Earthquake Occurrences Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Earthquakes')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

# 3D scatter plot for magnitude vs. depth
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(earthquake_data['source_magnitude'], earthquake_data['source_depth_km'], 
           earthquake_data['source_distance_km'], c=earthquake_data['source_magnitude'], cmap='viridis')
ax.set_xlabel('Magnitude')
ax.set_ylabel('Depth (km)')
ax.set_zlabel('Distance (km)')
plt.title('Magnitude vs Depth vs Distance')
plt.show()
```

```python
# Import necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Scatter plot for earthquake depth vs. distance
plt.figure(figsize=(10, 6))
sns.scatterplot(x='source_depth_km', y='source_distance_km', data=earthquake_data)
plt.title('Earthquake Depth vs. Source Distance')
plt.xlabel('Depth (km)')
plt.ylabel('Source Distance (km)')
plt.show()
```

```python
# Select columns relevant to seismic factors
seismic_factors = earthquake_data[['source_magnitude', 'source_depth_km', 
                                   'source_distance_km', 'snr_db', 'source_horizontal_uncertainty_km']]

# Compute correlation matrix
correlation_matrix = seismic_factors.corr()

# Plot heatmap of correlation matrix
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Between Seismic Factors')
plt.show()
```

```python
# Create magnitude ranges (e.g., small: 0-3, medium: 3-6, large: 6+)
earthquake_data['magnitude_range'] = pd.cut(earthquake_data['source_magnitude'], bins=[0, 3, 6, 10], labels=['Small', 'Medium', 'Large'])

# Plot the boxplot of source distance by magnitude range
plt.figure(figsize=(10, 6))
sns.boxplot(x='magnitude_range', y='source_distance_km', data=earthquake_data)
plt.title('Source Distance Across Different Magnitude Ranges')
plt.xlabel('Magnitude Range')
plt.ylabel('Source Distance (km)')
plt.show()
```

```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature

# Create a figure and add a map projection (PlateCarree projection is good for global maps)
plt.figure(figsize=(12, 8))
ax = plt.axes(projection=ccrs.PlateCarree())

# Add coastlines for the map
ax.coastlines()

# Add land and ocean features
ax.add_feature(cfeature.LAND, facecolor='lightgray')
ax.add_feature(cfeature.OCEAN)

# Plot earthquake locations (source_latitude and source_longitude)
plt.scatter(earthquake_data['source_longitude'], earthquake_data['source_latitude'], 
            c=earthquake_data['source_magnitude'], cmap='Reds', s=20, alpha=0.7, 
            transform=ccrs.PlateCarree())

# Add a colorbar to represent the magnitude of the earthquakes
plt.colorbar(label='Magnitude')

# Add title and labels
plt.title('Earthquake Locations on World Map (2015 Onwards)')
plt.xlabel('Longitude')
plt.ylabel('Latitude')

# Display theÂ plot
plt.show()
```

```python
# Scatter plot for earthquake depth vs. magnitude
plt.figure(figsize=(10, 6))
plt.scatter(earthquake_data['source_depth_km'], earthquake_data['source_magnitude'], 
            c='blue', alpha=0.7)
plt.title('Earthquake Depth vs Magnitude')
plt.xlabel('Depth (km)')
plt.ylabel('Magnitude')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

# 3D scatter plot for latitude, longitude, and depth
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
sc = ax.scatter(earthquake_data['source_longitude'], earthquake_data['source_latitude'], 
                earthquake_data['source_depth_km'], c=earthquake_data['source_magnitude'], cmap='viridis')

ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')
ax.set_zlabel('Depth (km)')
plt.title('3D Scatter Plot of Earthquake Locations and Depth')

# Add a colorbar for magnitudes
plt.colorbar(sc, label='Magnitude')

plt.show()
```
